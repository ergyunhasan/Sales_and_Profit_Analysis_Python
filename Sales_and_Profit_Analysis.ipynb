## Exploratory Data Analysis

### 1.What is the sales and the profit trend over time?

```python
superstore.head()
```

```python
superstore["order_date"].min()
```

```python
superstore["order_date"].max()
```

```python
# Creating a new column : year_month

superstore=superstore.assign(
                  year_month = superstore["order_date"].dt.strftime("%Y-%m")
).copy()
```

```python
superstore.head()
```

```python
sales_profit = superstore.groupby("year_month").agg({"sales":"sum", "profit":"sum"}).round(2).copy()
```

```python
sales_profit.head(12) 
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(
    sales_profit.index,
    sales_profit["sales"],
    color="#42b0f5",
    linewidth=3)

ax.set_title('Sales 2011-2014', fontsize=16)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Sales', fontsize=14)

ax.set_xticks(sales_profit.index[::3]);
plt.xticks(rotation=45);
```

The sales trend appears to be somewhat inconsistent, with some months showing increases and others showing decreases. However, there is an overall upward trend in sales from January through December.

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(
    sales_profit.index,
    sales_profit["profit"],
    color="#e31b64",
    linewidth=3)

ax.set_title('Profit 2011-2014', fontsize=16)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Profit', fontsize=14)

ax.set_xticks(sales_profit.index[::3]);
plt.xticks(rotation=45);
```

As for the profit trend, it also shows some variation from month to month, but it generally follows the same overall upward trend as sales, indicating that the company is generating more profit as sales increase.

### 2.What are the top 10 products by sales and by profit?

```python
superstore.head()
```

```python
# Top 10 products by sales

superstore.groupby(["product_name", "category", "sub_category"]) \
.agg({"sales":"sum","quantity":"sum"}) \
.sort_values("sales", ascending=False).reset_index().iloc[:10, :]
```

```python
# Top 10 products by profit

superstore.groupby(["product_name", "category", "sub_category"]) \
.agg({"profit":"sum", "quantity":"sum"}) \
.sort_values("profit", ascending=False).reset_index().iloc[:10, :]
```

- When comparing Top 10 products by sales and Top 10 products by profit that some of the products are not in the both list such as Apple Smart Phone, Full Size is the first product on the top 10 products by sales, but it is not on Top 10 products by profit. The quantity of this product is more than any product on both list.


- Technology is the most important category for the products on both tables.


- Phone is the most important sub_category for the products on both tables.

### 6.Which is the most preferred ship mode?

```python
shipping = superstore \
.groupby("ship_mode") \
.agg({"order_id":"count"}) \
.sort_values("order_id", ascending=False)
```

```python
shipping
```

```python
ax = sns.barplot(
            x=shipping.index,
            y=shipping.order_id,
            data= shipping
)

ax.set_title("Shipping Mode Categories")
ax.set_xlabel("Shipping Mode")
ax.set_ylabel("Count");
```

### 7.What are the top countries and markets by sales ?

```python
top10countries = superstore \
.groupby(["market","country"], as_index=False) \
.agg({"sales":"sum"}).round(2) \
.sort_values("sales", ascending=False) \
.iloc[:10, :].reset_index(drop=True)
```

```python
top10countries
```

```python
fig, ax = plt.subplots(figsize=(11,7))

ax = sns.barplot(
    
    x=top10countries.country,
    y=top10countries.sales,
    data=top10countries
)

ax.set_title("Top 10 Countries by Sales", fontsize=14)
ax.set_xlabel("Country", fontsize=12)
ax.set_ylabel("Sales (Millions)", fontsize=12);
```

**United States** is the top country regarding sales and there is a significant difference with other markets.

```python
sales_markets = superstore \
.groupby("market", as_index=False) \
.agg({"sales":"sum"}).round(2) \
.sort_values("sales", ascending=False).reset_index(drop=True)
```

```python
sales_markets
```

```python
profit_markets = superstore \
.groupby("market", as_index=False) \
.agg({"profit":"sum"}).round(2) \
.sort_values("profit", ascending=False).reset_index(drop=True)
```

```python
profit_markets
```

```python
figure, axis = plt.subplots(1, 2, figsize=(12,5))

ax = sns.barplot(
    
    x=sales_markets.sales,
    y=sales_markets.market,
    data=sales_markets,
    ci=None,
    ax=axis[0]
)

ax.set_title("Markets by Sales", fontsize=14)
ax.set_xlabel("Sales (Millions)", fontsize=12)
ax.set_ylabel("Market", fontsize=12);


ax = sns.barplot(
    
    x=profit_markets.profit,
    y=profit_markets.market,
    data=profit_markets,
    ci=None,
     ax=axis[1]
)

ax.set_title("Markets by Profit", fontsize=14)
ax.set_xlabel("Profit", fontsize=12)
ax.set_ylabel("Market", fontsize=12)

figure.tight_layout();
```

- **The APAC** region had the highest sales figure of over 3.5 million dollars, as well as the highest profit figure of over 437 thousand dollars, indicating strong performance in terms of both sales and profitability. **The EU and US** regions also had relatively high sales and profit figures, indicating strong performance in these regions as well.


- **The Canada** region had relatively low sales and profit figures, indicating that this region may not be as profitable compared to other regions. Similarly, **the Africa and EMEA** regions had relatively low sales and profit figures, suggesting that these regions may require additional attention to improve their performance.
